step 1- importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

step 2- importing dataset

```python
data= pd.read_csv("data.csv")
```

```python
data.head() 
```

step 3- plot graphs (Bi-variate Analysis)

```python
data.plot(x='Hours',y='Scores',style="1")
plt.title("Hours VS Scores")
plt.xlabel("hours studies")
plt.ylabel("score")
plt.show()
```

```python
data.plot.pie(x="Hours",y="Scores")
```

```python
data.plot.scatter(x="Hours",y="Scores")
```

```python
data.plot.bar(x="Hours",y="Scores")
```

```python
data.sort_values(["Hours"],axis=0,ascending=[True],inplace=True)
data.head(10)
data.plot.bar(x="Hours",y="Scores")
```

After plotting several graphs, we observe that there is a linear relationship between the two variables.

step 4- Prepare data for model

```python
x=data.iloc[:,:-1].values
y=data.iloc[:,1].values
print(x)
print(y)
```

step 5- Divide the data for training and testing the model

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

step 6- Training Model

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
print("training complete")
```

```python
print("actual value")
print(y_test)
y_pred=regressor.predict(x_test)
print("predicted value")
print(y_pred)
```

step 7- Check the accuracy of the model

```python
df=pd.DataFrame({"actual":y_test,"predicted":y_pred})
df
```

step 8- Prediction with custom input

```python
hours=[[7]]
pred=regressor.predict(hours)
print(pred)
```

step 9- Evaluating the model

```python
from sklearn import metrics 
print("mean absolute error:",metrics.mean_absolute_error(y_test,y_pred))
```
